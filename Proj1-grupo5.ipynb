## 3. Building our Triple-Store

### Importing the libraries and transforming the data

As you can see in the next cells we used the rdflib package to build our triple-store and we used pandas to structure the data that we want to use.

```python
#import labraries
from rdflib import Graph, URIRef, RDF, Literal, BNode,Namespace
from rdflib.namespace import FOAF, XSD, RDFS
import pandas
import numpy as np
```

```python
#import files and create two dataframes for each file
lines=pandas.read_csv("insects.csv",sep=";", encoding="utf-8")
insects_csv= pandas.DataFrame(lines)

#slice the columns of insects data
insects_csv=insects_csv[['Unique Entry ID', 'Name', 'Sell', 'Where/How', 'Weather','Total Catches to Unlock', 'Spawn Rates', 'NH Jan', 'NH Feb', 'NH Mar','NH Apr', 'NH May', 'NH Jun', 'NH Jul', 'NH Aug', 'NH Sep', 'NH Oct',
       'NH Nov', 'NH Dec','Color 1', 'Color 2']]
insects_csv=insects_csv.replace(np.nan)
#create dictionaries from the data
insectos = insects_csv.to_dict('index')

#append each line of the dictionary into an array
insects_array=[]
for i in insectos:
    insects_array.append(insectos[i])


lines2=pandas.read_csv("fish.csv",sep=";",encoding="utf-8")
fish_csv= pandas.DataFrame(lines2)
columns_fish=fish_csv.columns


fish_csv=fish_csv[['Unique Entry ID', 'Name', 'Sell', 'Where/How', 'Shadow','Total Catches to Unlock', 'Rain/Snow Catch Up', 'Spawn Rates', 'NH Jan', 'NH Feb', 'NH Mar','NH Apr', 'NH May', 'NH Jun', 'NH Jul', 'NH Aug', 'NH Sep', 'NH Oct',
       'NH Nov', 'NH Dec','Color 1', 'Color 2', 'Lighting Type']]
fish_csv=fish_csv.replace(np.nan)

peixes = fish_csv.to_dict('index')
fish_array=[]
for p in peixes:
    fish_array.append(peixes[p])
```

```python
#Create the namespace for animal crossing 
G=Graph()

ac = Namespace('http://www.di.fc.ul.pt/~Carolina_João/Animal_crossing#')

likesWater = URIRef("likesWater")
G.add((likesWater, RDF.type, XSD.boolean))

likesPlants = URIRef("likesPlants")
G.add((likesPlants, RDF.type, XSD.boolean))

canFly = URIRef("canFly")
G.add((canFly, RDF.type, XSD.boolean))

#create set of unique products

for i in insects_array:
    #take special note of the next line! - this is how we append a namespace to a string to make a URI
    G.bind("ac",ac)
    insecto_name = URIRef(i["Name"])
    ueid= URIRef(ac+i["Unique Entry ID"])
    sell_price=Literal(i["Sell"],datatype=XSD.int)
    where_how= URIRef(i["Where/How"])
    weather= URIRef(i["Weather"])
    tcatches2unlock=Literal(i["Total Catches to Unlock"],datatype=XSD.int)
    catches2unlock= Literal(i["Total Catches to Unlock"], datatype=XSD.int)
    spawn= Literal(i["Spawn Rates"])
    hours_jan= Literal(i["NH Jan"])
    hours_feb= Literal(i["NH Feb"])
    hours_mar= Literal(i["NH Mar"])
    hours_apr= Literal(i["NH Apr"])
    hours_may= Literal(i["NH May"])
    hours_jun= Literal(i["NH Jun"])
    hours_jul= Literal(i["NH Jul"])
    hours_aug= Literal(i["NH Aug"])
    hours_sep= Literal(i["NH Sep"])
    hours_oct= Literal(i["NH Oct"])
    hours_nov= Literal(i["NH Nov"])
    hours_dec= Literal(i["NH Dec"])
    color1= Literal(i["Color 1"])
    color2= Literal(i["Color 2"])

    
   
    G.add((ueid, RDF.type, ac.Insect))
    G.add((ueid, RDFS.label, Literal(insecto_name, lang="en")))
    G.add((ueid, ac.where_how,Literal(where_how)))
    G.add((ueid, ac.sell_price,Literal(sell_price,datatype=XSD.int)))
    G.add((ueid, ac.catches_to_unlock,Literal(tcatches2unlock,datatype=XSD.int)))
    G.add((ueid, ac.spawn,Literal(spawn)))
    G.add((ueid, ac.hours_jan,Literal(hours_jan)))
    G.add((ueid, ac.hours_feb,Literal(hours_feb)))
    G.add((ueid, ac.hours_mar,Literal(hours_mar)))
    G.add((ueid, ac.hours_apr,Literal(hours_apr)))
    G.add((ueid, ac.hours_may,Literal(hours_may)))
    G.add((ueid, ac.hours_jun,Literal(hours_jun)))
    G.add((ueid, ac.hours_jul,Literal(hours_jul)))
    G.add((ueid, ac.hours_aug,Literal(hours_aug)))
    G.add((ueid, ac.hours_sep,Literal(hours_sep)))
    G.add((ueid, ac.hours_oct,Literal(hours_oct)))
    G.add((ueid, ac.hours_nov,Literal(hours_nov)))
    G.add((ueid, ac.hours_dec,Literal(hours_dec)))
    G.add((ueid, ac.color1,Literal(color1)))
    G.add((ueid, ac.color2,Literal(color2)))
    
    
    if ("water" in i["Where/How"])or ("ponds" in i["Where/How"]) or ("rivers" in i["Where/How"]):
        G.add((ueid, ac.likesWater,Literal(True)))
    
    if "flowers" or "turnips" or "trees" or "tree"  in i["Where/How"]:
        G.add((ueid, ac.likesPlants,Literal(True)))
        
    if "Flying" in i["Where/How"]:
        G.add((ueid, ac.canFly,Literal(True)))
    
    
    G.add((ueid, ac.weather,Literal(weather)))

    #and here we give it a label (labels are human readable, URIs, may not be so)
 
print(G.serialize(format='n3'))
#print(G.serialize(format='turtle').decode('utf-8'))
```

```python
for f in fish_array:
    #take special note of the next line! - this is how we append a namespace to a string to make a URI
    G.bind("ac",ac)
    ueid= URIRef(ac+f["Unique Entry ID"])
    
    saltWater = URIRef("saltWater")
    G.add((saltWater, RDF.type, XSD.boolean))
    
    sweetWater = URIRef("sweetWater")
    G.add((sweetWater, RDF.type, XSD.boolean))
    
    name = Literal(f["Name"])
    sell_price=Literal(f["Sell"],datatype=XSD.int)
    where_how= URIRef(f["Where/How"])
    shadow= Literal(f["Shadow"])
    catches2unlock= Literal(f["Total Catches to Unlock"], datatype=XSD.int)
    spawn= Literal(f["Spawn Rates"])
    rain_snow= Literal(f["Rain/Snow Catch Up"])
    hours_jan= Literal(f["NH Jan"])
    hours_feb= Literal(f["NH Feb"])
    hours_mar= Literal(f["NH Mar"])
    hours_apr= Literal(f["NH Apr"])
    hours_may= Literal(f["NH May"])
    hours_jun= Literal(f["NH Jun"])
    hours_jul= Literal(f["NH Jul"])
    hours_aug= Literal(i["NH Aug"])
    hours_sep= Literal(f["NH Sep"])
    hours_oct= Literal(f["NH Oct"])
    hours_nov= Literal(f["NH Nov"])
    hours_dec= Literal(f["NH Dec"])
    color1= Literal(f["Color 1"])
    color2= Literal(f["Color 2"])
    lighting= Literal(f["Lighting Type"])
    

    G.add((ueid, RDF.type, ac.Fish))
    G.add((ueid, RDFS.label, Literal(name, lang="en")))
    G.add((ueid, ac.sell_price,Literal(sell_price, datatype=XSD.int)))
    G.add((ueid, ac.where_how,Literal(where_how)))
    G.add((ueid, ac.shadow,Literal(shadow)))
    G.add((ueid, ac.catches2unlock,Literal(catches2unlock, datatype=XSD.int)))
    G.add((ueid, ac.spawn,Literal(spawn)))
    G.add((ueid, ac.rain_snow,Literal(rain_snow)))
    G.add((ueid, ac.hours_jan,Literal(hours_jan)))
    G.add((ueid, ac.hours_feb,Literal(hours_feb)))
    G.add((ueid, ac.hours_mar,Literal(hours_mar)))
    G.add((ueid, ac.hours_apr,Literal(hours_apr)))
    G.add((ueid, ac.hours_may,Literal(hours_may)))
    G.add((ueid, ac.hours_jun,Literal(hours_jun)))
    G.add((ueid, ac.hours_jul,Literal(hours_jul)))
    G.add((ueid, ac.hours_aug,Literal(hours_aug)))
    G.add((ueid, ac.hours_sep,Literal(hours_sep)))
    G.add((ueid, ac.hours_oct,Literal(hours_oct)))
    G.add((ueid, ac.hours_nov,Literal(hours_nov)))
    G.add((ueid, ac.hours_dec,Literal(hours_dec)))
    G.add((ueid, ac.color1,Literal(color1)))
    G.add((ueid, ac.color2,Literal(color2)))
    G.add((ueid, ac.lighting,Literal(lighting)))
    
    G.add((ueid, ac.likesWater,Literal(True)))
    
    if 'Sea' in f["Where/How"]:
        G.add((ueid, ac.saltWater,Literal(True)))  
    else:
        G.add((ueid, ac.sweetWater,Literal(True)))
    
        


    #and here we give it a label (labels are human readable, URIs, may not be so)
    
print(G.serialize(format='turtle'))
#print(G.serialize(format='turtle').decode("utf-8"))
```

## 4. Queries

### Query 1 - Insects that like being near water.

```python
q1 = G.query(
    """SELECT ?name
       WHERE {
          ?i a ac:Insect .
          ?i rdfs:label ?name.
          ?i ac:likesWater true .
       }""")

for row in q1:
    print("Name: %s" % row)
```

### Query 2 - Fishes that live in sweet water.

```python
q2 = G.query(
    """SELECT ?name
       WHERE {
          ?f a ac:Fish .
          ?f rdfs:label ?name.
          ?f ac:sweetWater True .
       }""")

for row in q2:
    print("Name: %s" % row)
```

### Query 3 - Animals that we can find in or near watersource.

```python
q3 = G.query(
    """SELECT ?name
       WHERE{
       {
          ?i a ac:Insect .
          ?i rdfs:label ?name.
          ?i ac:likesWater true .}
          UNION{
              ?f a ac:Fish .
              ?f rdfs:label ?name.
              ?f ac:likesWater true .
          }
       }
    """)
       

for row in q3:
    print("Name: %s" % row)
```
